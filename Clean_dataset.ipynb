```python
import pandas as pd

# Cargar dataset
df = pd.read_csv("dataset_anemia_2020_2024.csv")

# Seleccionar únicamente las columnas necesarias
df = df[[
    "ID1",
    "HW1",
    "HW2",
    "HW3",
    "HW53",
    "HW70",
    "HW71",
    "HW72",
    "HW73",
    "HW57"
]]

# Renombrar columnas
df.rename(columns={
    "ID1": "ANO",
    "HW1": "Edad_Meses",
    "HW2": "Peso_kg",
    "HW3": "Talla_cm",
    "HW53": "Hemoglobina",
    "HW70": "Zscore_Talla_Edad",
    "HW71": "Zscore_Peso_Edad",
    "HW72": "Zscore_Peso_Talla",
    "HW73": "Zscore_IMC",
    "HW57": "Nivel_Anemia"
}, inplace=True)

# Convertir espacios en blanco en NaN
df.replace(r'^\s*$', pd.NA, regex=True, inplace=True)
df = df.dropna()

# Contar valores nulos (NaN)
nulos = df.isna().sum()

# Contar espacios en blanco ("" o "   ")
espacios = df.apply(lambda col: col.astype(str).str.strip().eq("").sum())

# Crear tabla resumen
resumen = pd.DataFrame({
    "Valores Nulos": nulos,
    "Espacios en Blanco": espacios,
    "Total Faltantes": nulos + espacios
})

print(resumen)

df["Nivel_Anemia"] = pd.to_numeric(df["Nivel_Anemia"], errors="coerce")
cantidad = (df["Nivel_Anemia"] >= 5).sum()

print(f"Cantidad de registros con Nivel_Anemia >= 5: {cantidad}")

df_final = df[df["Nivel_Anemia"] < 5]
cantidad = (df_final["Nivel_Anemia"] >= 5).sum()

print(f"Cantidad de registros con Nivel_Anemia >= 5 despues de eliminar: {cantidad}")

df_final.to_csv("dataset_final.csv", index=False)
```
