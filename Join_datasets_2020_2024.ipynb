```python
import pandas as pd
from google.colab import files

# Cargar los datasets
df1 = pd.read_csv('REC44_2020.csv')
df2 = pd.read_csv('REC44_2021.csv')
df3 = pd.read_csv('REC44_2022.csv')
df4 = pd.read_csv('REC44_2023.csv')
df5 = pd.read_csv('REC44_2024.csv')

# Juntar verticalmente
df_unido = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

# Guardar el df en un nuevo archivo CSV
df_unido.to_csv('dataset_anemia_2020_2024.csv', index=False)

# Descargar automática
files.download('dataset_anemia_2020_2024.csv')
```
